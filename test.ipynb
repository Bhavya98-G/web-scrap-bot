```python
from webscrap import extract_meaningful_text
from text_processing import TextChunker
from embedding import VectorStoreManager
from ai_handler import answer_question
```

```python
text = extract_meaningful_text("https://en.wikipedia.org/wiki/Web_scraping")
print(text)
```

```python
chunker = TextChunker(chunk_size=50, chunk_overlap=10,unit="words")
chunks = chunker.process(text["content"])
chunks
```

```python
vector_store = VectorStoreManager()
vector_store.build_index(chunks)
```

```python
result = answer_question("What was my last question?", vector_store, "https://en.wikipedia.org/wiki/Web_scraping", "Web Scraping")
```
